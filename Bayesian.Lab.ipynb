```python
import pandas as pd
import numpy as np
from scipy.stats import beta, gamma
import matplotlib.pyplot as plt
```

```python
ay, by, a0, b0 = 9, 11, 2, 3
A, B = (a0, a0 + ay),(b0, b0 + by)
M = (A[0] - 1) / (A[0] + B[0] - 2), (A[1] - 1) / (A[1] + B[1] - 2)
X = np.linspace(0, 1, 1001)
Y = [beta.pdf(X, a, b) for a, b in zip(A, B)]

plt.gcf().set_size_inches(8, 8)
plt.plot(X, Y[0], label='Prior Beta({}, {})'.format(A[0], B[0]))
plt.plot(X, Y[1], label='Posterior Beta({}, {})'.format(A[1], B[1]))
plt.plot([M[0], M[0]], [0, max(Y[1]) + 0.2], linestyle='dashed', 
         label='Prior Mode = {:.3f}'.format(M[0]))
plt.plot([M[1], M[1]], [0, max(Y[1]) + 0.2], linestyle='dashed', 
         label='Posterior Mode  = {:.3f}'.format(M[1]))
plt.xlim(0, 1)
plt.grid(axis='both')
plt.ylabel('PDF (%)')
plt.xlabel('Theta')
plt.legend(loc='best')
plt.title('Prior and Posterior Densities of Theta with Y=9, N-Y=11')
;
```
